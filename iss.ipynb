```python
import requests
import matplotlib.pyplot as plt
import datetime
import time

def location():
    loc = requests.get('http://api.open-notify.org/iss-now.json')
    data = loc.json()
    time = datetime.datetime.fromtimestamp(data['timestamp'])
    latitude = float(data['iss_position']['latitude'])
    longitude = float(data['iss_position']['longitude'])
    return latitude, longitude, time

points = []
for each in range(100):
    latitude, longitude, timestamp = location()
    points.append((latitude, longitude, time))
    time.sleep(10)
    
```

```python
latitudes = [point[0] for point in points]
longitudes = [point[1] for point in points]

# Plotting the path of ISS
plt.plot(longitudes, latitudes)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Path of International Space Station')
plt.grid(True)
plt.show()
```
